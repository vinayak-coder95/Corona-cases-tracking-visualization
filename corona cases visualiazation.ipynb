```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
data.head()
```

```python
df = data.iloc[:-1,1:]
```

```python
df.head()
```

```python
Total_states = np.arange(len(df['state_name']))
```

```python
Total_states
```

```python
from matplotlib.pyplot import figure
plt.figure(figsize=(10,10))
plt.barh(Total_states,df['positive'])
    
plt.yticks(Total_states, df['state_name'])  
plt.xlim(1,max(df['positive'])+100) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
df = df.set_index('state_name',drop=True)
```
